```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer=RegexpTokenizer(r'[A-Za-z]+')
```

```python
df.URL[0]
```

```python
tokenizer.tokenize(df.URL[0])
```

```python
df['text_tokenized']=df.URL.map(lambda t: tokenizer.tokenize(t))
```

```python
df.head()
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer=SnowballStemmer('english')
```

```python
df['text_stemmed']=df['text_tokenized'].map(lambda l: [stemmer.stem(word) for word in l])
```

```python
df.head()
```

```python
df['text']=df['text_stemmed'].map(lambda l: ' '.join(l))
```

```python
df.head()
```

```python
good_sites=df[df.Label == 'good']
bad_sites=df[df.Label == 'bad']
```

```python
def plot_wordcloud(text, mask=None, max_words=400, max_font_size=120, figure_size=(24.0, 16.0),
                   title=None, title_size=40, image_color=False):

    stopwords = set(STOPWORDS)
    more_stopwords = {'coo', 'http'}
    stopwords = stopwords.union(more_stopwords)

    wordcloud = WordCloud(background_color='white',
                          stopwords=stopwords,
                          max_words=max_words,
                          max_font_size=max_font_size,
                          random_state=42,
                          mask=mask)

    wordcloud.generate(text)

    plt.figure(figsize=figure_size)

    if image_color:
        image_colors = ImageColorGenerator(mask)
        plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
        plt.title(title, fontdict={'size': title_size,
                                   'verticalalignment': 'bottom'})
    else:
        plt.imshow(wordcloud)
        plt.title(title, fontdict={'size': title_size, 'color': 'green',
                                   'verticalalignment': 'bottom'})

    plt.axis('off')
    plt.tight_layout()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
features = cv.fit_transform(df.text)
```

```python
features[:5].toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, df.Label,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
l_model = LogisticRegression()
```

```python
l_model.fit(x_train, y_train)
```

```python
l_model.score(x_test, y_test)
```

```python
l_model.score(x_train, y_train)
```

```python
from sklearn.metrics import classification_report

print('\nCLASSIFICATION REPORT\n')
print(classification_report(l_model.predict(x_test), y_test, 
                            target_names=['Bad', 'Good']))
```
